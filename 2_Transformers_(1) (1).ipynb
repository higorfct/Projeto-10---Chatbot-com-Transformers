#Criando o chatbot com Transformers

```python
import transformers
from transformers import pipeline
```

```python
qea = pipeline("question-answering", model="pierreguillou/bert-base-cased-squad-v1.1-portuguese")
```

```python
texto = (
    "O cartão Platinum oferece uma série de benefícios exclusivos aos seus usuários. "
    "Além de um limite de crédito que pode chegar a até 20 mil reais, ele proporciona acesso a salas VIP em aeroportos, "
    "programas de pontos que podem ser trocados por produtos ou viagens, e isenção da anuidade nos primeiros 12 meses após a adesão. "
    "O cartão é indicado para consumidores que buscam vantagens em viagens e compras."
)
pergunta = "Qual o limite de crédito do cartão Platinum?"
resposta = qea(question=pergunta, context=texto)
print("Pergunta: ", pergunta)
print("Resposta: ", resposta['answer'])
print("Score: ", resposta['score'])
```

#Gerando uma aplicação web com Gradio

```python
from transformers import pipeline
import gradio as gr

# Texto fixo como contexto
contexto = (
    "O cartão Platinum oferece uma série de benefícios exclusivos aos seus usuários. "
    "Além de um limite de crédito que pode chegar a até 20 mil reais, ele proporciona acesso a salas VIP em aeroportos, "
    "programas de pontos que podem ser trocados por produtos ou viagens, e isenção da anuidade nos primeiros 12 meses após a adesão. "
    "O cartão é indicado para consumidores que buscam vantagens em viagens e compras."
)

# Carrega o pipeline de QA em português
qa = pipeline("question-answering", model="pierreguillou/bert-base-cased-squad-v1.1-portuguese")

def responder(pergunta):
    resultado = qa(question=pergunta, context=contexto)
    return resultado['answer']

# Interface Gradio
iface = gr.Interface(
    fn=responder,
    inputs=gr.Textbox(label="Digite sua pergunta"),
    outputs=gr.Textbox(label="Resposta"),
    title="Chatbot QA - Cartão Platinum",
    description="Faça perguntas sobre os benefícios do cartão Platinum com base no texto fixo."
)

# Rodar no notebook com interface embutida e link público
iface.launch(share=True, inline=True)
```
